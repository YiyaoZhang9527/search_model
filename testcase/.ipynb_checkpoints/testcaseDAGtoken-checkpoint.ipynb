# 加载概率字典，没有的词假定一个极小值，比如10**(-8)

```python
def freq_function(loadpath = "dict"):
    dict_file = open(loadpath,'r').readlines()
    m = len(dict_file)
    freq = dict()
    maxmum_lenght = 0
    for row in dict_file:
        rowlist = row.split()
        word , freq_counter = rowlist[0],rowlist[1]
        """log处理极小值问题，加负号将问题从最大值问题转为最小值问题"""
        freq_var = -math.log(int(freq_counter)/m)
        freq.update({word:freq_var})
        temp_lenght = len(word)
        if temp_lenght>maxmum_lenght:
            maxmum_lenght = temp_lenght
    return freq,maxmum_lenght

FREQ =freq_function()
FREQ[-1]
```

# 关联图(前向最大匹配)

```python

def forward_node(sentence,dict_):
    m = dict_[-1]
    freq = dict_[0]
    n = len(sentence)
    #循环句子内的字符串，如果出现了，就记录下出现的每一个组合的词的长度下标
    r = dict()
    for i in range(n):
        for j in range(i,n):
            temp_strings = sentence[i:j]
            if temp_strings in freq:
                if freq[temp_strings]:
                    if i not in r:
                        r.update({i:{j}})
                    else:
                        r[i].update({j})        
            else:
                if i not in r:
                    r.update({i:{i}})
                else:
                    r[i].update({i})
    return r

forward_node(case,freq_function())
```

# 概率图

```python
def path_probability_graph(sentence,freq=freq_function()):
    return {i:{z:freq[0][case[i:z]]  if case[i:z] in freq[0] else 10**-8 for z in j} for i,j in forward_node(sentence,freq).items()}
path_probability_graph(case)
```

```python
plt.figure(figsize=(16,9))
for i,iter_ in path_probability_graph(case).items():
    for j,p in iter_.items():
        if p != 10**-8:
            plt.scatter(i,p)
            plt.annotate((str(i),str(j)),xy=(i,p),xycoords='data',xytext=(+30,-30),
            textcoords='offset points',arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=.2"))
        #plt.plot(range(i,j),list(range(i,j)))
plt.show()
```

```python
def probability_graph_maxtrix(sentence,freq=freq_function()):
    rdict = dict()
    for i,iter_j in forward_node(sentence,freq).items():
        for j in iter_j:
            token = sentence[i:j]
            if token in freq[0]:
                rdict.update({token:freq[0][token]})
            else:
                rdict.update({token:10**-8})
    return rdict

probability_graph_maxtrix(case)
```

```python
def dag_funtion(sentence,FREQ=freq_function()[0]):
        rs = {}    
        n = len(sentence) 
        for k in range(n): 
            temp = [] 
            i = k
            temp_token = sentence[k] 
            while i < n and temp_token in FREQ:
                if FREQ[temp_token]: 
                    temp.append(i)  
                i += 1     
                temp_token = sentence[k:i + 1]  
            if not temp:  
                temp.append(k)
            rs[k] = temp 
        return rs

dag_funtion(case)
```

# 利用DAG分词

```python
def token(original):
    han= "".join(re.split('^(.+?)( [0-9]+)?( [a-z]+)?$',original))
    DAG=dag_funtion(han)
    rl = []
    for i,index_s in DAG.items():
        temp = []
        for index_ in index_s:
            temp.append(original[index_])
        rl.append("|"+"".join(temp))
    return "".join(rl)
token(case)
```
